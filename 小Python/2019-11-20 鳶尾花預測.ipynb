```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# 載入鳶尾花數據

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
x[:5]
```

```python
y[:5]
```

# x選用後兩個條件

```python
X = x[:,2:]
Y = y
```

# 區分訓練資料和測試資料

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train ,y_test = train_test_split(X , Y , test_size = 0.2 , random_state = 0)
```

# 使用SVM

```python
from sklearn.svm import SVC
```

```python
clf = SVC(gamma='auto')
```

```python
clf.fit(x_train,y_train)
```

```python
y_predict = clf.predict(x_test)
```

# 用y_oredict扣掉y_test之後畫圖看到預測命中率非常高

```python
plt.scatter(x_test[:,0] , x_test[:,1] , c=y_predict-y_test)
```

# 畫圖做圖形辨識看到準確率非常高

```python
x1 = np.arange(0.5,7.2,0.02)
y1 = np.arange(0,3,0.02)
x2,y2 = np.meshgrid(x1,y1)
P = np.c_[x2.ravel() , y2.ravel()]
z = clf.predict(P)
Z = z.reshape(x2.shape)
plt.contourf(x2 , y2 , Z , cmap='Paired' , alpha=0.3)
plt.scatter(x_test[:,0] , x_test[:,1] , cmap='Paired' , c=y_test)
```

# 結論:換成後面2個條件之後非常準確
